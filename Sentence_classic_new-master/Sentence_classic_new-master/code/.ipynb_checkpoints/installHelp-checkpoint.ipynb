```python
import csv
import re
with open('C:/Users/Aravindhan.Poopathy/OneDrive/Arav/Sentence_classic_new-master/Sentence_classic_new-master/code/data/twitter_100000.csv', newline='') as File:
    reader = csv.reader(File)
    trainData = []
    for row in reader:
        row1 = row[1]
        row2 = re.sub(r'[?|$|.|!|#|*]', '', row[2])
        row2 = " ".join(filter(lambda x:x[0] not in ['@','!','#','*','&'], row2.split()))
        row = row1+ "\t" +row2.strip()
        trainData.append(row)
del trainData[0]
trainData
```

```python
trainData1 = [i for i in trainData if i[0]=='0']
# # row
# text = '@viju I am good thanks!'
# j = '1'
# # for i in text:
# #     if i.startswith('@'):
# #         continue
# #     else:
# #         j = j +

# import re
# text = re.sub(r'@', '', text)
# text
trainData2 = [i for i in trainData if i[0]=='1']
final_TrainData = trainData1 + trainData2
len(final_TrainData)
```

```python
def writeFile(filename, sentences):
    with open(filename, 'w') as fOut:
        for line in sentences:
            fOut.write(line)
            fOut.write("\n")

writeFile("C:/Users/Aravindhan.Poopathy/OneDrive/Arav/Sentence_classic_new-master/Sentence_classic_new-master/code/data/twitterTrianData.txt", final_TrainData)
```

```python
from random import shuffle
shuffle(final_TrainData)
    
n = len(final_TrainData)    
trainSplit = (0, int(0.5*n))
devSplit = (trainSplit[1], trainSplit[1]+int(0.25*n))
testSplit = (devSplit[1], n)

train = final_TrainData[trainSplit[0]:trainSplit[1]]
dev = final_TrainData[devSplit[0]:devSplit[1]]
test = final_TrainData[testSplit[0]:testSplit[1]]

writeFile('C:/Users/Aravindhan.Poopathy/OneDrive/Arav/Sentence_classic_new-master/Sentence_classic_new-master/code/twitterDatasplit/tweettrain.txt', train)
writeFile('C:/Users/Aravindhan.Poopathy/OneDrive/Arav/Sentence_classic_new-master/Sentence_classic_new-master/code/twitterDatasplit/tweetdev.txt', dev)
writeFile('C:/Users/Aravindhan.Poopathy/OneDrive/Arav/Sentence_classic_new-master/Sentence_classic_new-master/code/twitterDatasplit/tweettest.txt', test)
```
